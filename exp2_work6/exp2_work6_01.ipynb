```python
%matplotlib inline
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd

# function_define
def gaussian(x, a11, a12, a13):
    return a11*np.exp(-((x-a12)/a13)**2/2)

# data_input
xdata = np.linspace(1, 6, 4)
ydata = np.array([2, 4, 7, 5])

# prameter_initial
prameter_initial = np.array([1,2,3,])

# fitting
popt, pcov = curve_fit(gaussian, xdata, ydata, p0=prameter_initial)

y_ = gaussian(np.linspace(0,10), np.ones(50)*popt[0], np.ones(50)*popt[1], np.ones(50)*popt[2])

plt.plot(np.linspace(0,10), y_)

plt.scatter(xdata, ydata)
plt.show()
```
